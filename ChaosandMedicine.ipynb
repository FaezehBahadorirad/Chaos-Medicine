```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
# Replace the path with the actual path to the dataset downloaded from Kaggle
dataframe = pd.read_csv('/Users/faezehbahadorirad/Downloads/archive-6/mitbih_train.csv', header=None)
# Selecting a single ECG recording (for demonstration purposes, we pick the first one)
# Typically, each row is a separate ECG recording
ecg = dataframe.iloc[0, :-1]  # Exclude the last column which is the label

# Data normalization to fit within a [0, 1] range
scaler = MinMaxScaler()
ecg_normalized = scaler.fit_transform(ecg.values.reshape(-1, 1)).flatten()

# Plotting the ECG
plt.figure(figsize=(14, 6))
plt.plot(ecg_normalized)
plt.title('Normalized ECG Signal')
plt.xlabel('Time')
plt.ylabel('Normalized Amplitude')
plt.show()

# Simple analysis to demonstrate chaos theory concepts:
# Calculating the difference between consecutive ECG signal points
diff_ecg = np.diff(ecg_normalized)
# Plotting the difference
plt.figure(figsize=(14, 6))
plt.plot(diff_ecg)
plt.title('Difference Between Consecutive ECG Signal Points')
plt.xlabel('Time')
plt.ylabel('Difference in Normalized Amplitude')
plt.show()

# Simple Chaos Analysis
# Plotting a delay embedding to visualize potential chaotic patterns
plt.figure(figsize=(8, 8))
plt.scatter(ecg_normalized[:-1], ecg_normalized[1:], alpha=0.5)
plt.title('ECG Signal Delay Embedding')
plt.xlabel('ECG[n]')
plt.ylabel('ECG[n+1]')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(r, x):
    return r * x * (1 - x)

# Simulation settings
n_iterations = 100
r = 3.9  # Parameter value known to lead to chaotic behavior

# Initial conditions
x0_a = 0.5  # Initial condition for scenario A
x0_b = x0_a + 1e-5  # Initial condition for scenario B, slightly different from A

# Initializing arrays to store the simulation results
x_a = np.zeros(n_iterations)
x_b = np.zeros(n_iterations)

# Setting the first element of each array to the initial condition
x_a[0] = x0_a
x_b[0] = x0_b

# Running the simulation
for i in range(1, n_iterations):
    x_a[i] = logistic_map(r, x_a[i-1])
    x_b[i] = logistic_map(r, x_b[i-1])

# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(x_a, label='Scenario A', linewidth=2)
plt.plot(x_b, label='Scenario B', linewidth=2, linestyle='--')
plt.title('Butterfly Effect in a Chaotic System (Logistic Map) - Medicine')
plt.xlabel('Iteration')
plt.ylabel('Value')
plt.legend()
plt.show()
```
